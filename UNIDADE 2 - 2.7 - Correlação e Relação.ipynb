```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("penguins")
```

```python
df
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    s=80
)

# linha de tendência
sns.regplot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    scatter=False,
    color="black"
)

plt.title("Relação entre tamanho da nadadeira e peso")
plt.xlabel("Comprimento da nadadeira (mm)")
plt.ylabel("Massa corporal (g)")

plt.show()
```
