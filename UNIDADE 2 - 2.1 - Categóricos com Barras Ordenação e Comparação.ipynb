```python
import pandas as pd
import matplotlib.pyplot as plt

dados = {
    "Produto": ["A", "B", "C", "D"],
    "Receita": [120000, 85000, 43000, 150000]
}

df = pd.DataFrame(dados)
df = df.sort_values("Receita", ascending=False)
```

```python
df
```

```python
plt.bar(df["Produto"], df["Receita"])
plt.title("Receita por Produto")
plt.ylabel("Receita (R$)")

for i, receita in enumerate(df["Receita"]):
    plt.text(i, receita, f'{receita:,.2f}', ha='center', va='bottom')

plt.show()
```
