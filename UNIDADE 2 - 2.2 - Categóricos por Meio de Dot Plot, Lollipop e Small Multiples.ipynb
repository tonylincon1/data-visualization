# Dot plot

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Categoria": ["A", "B", "C", "D", "E"],
    "Valor": [10, 25, 15, 30, 18]
}).sort_values("Valor")

plt.figure(figsize=(7, 4))
plt.scatter(df["Categoria"], df["Valor"])
plt.title("Dot plot: comparação de categorias")
plt.ylabel("Valor")
plt.xlabel("Categoria")
plt.tight_layout()
plt.show()
```

# Lollipop

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Categoria": ["A", "B", "C", "D", "E"],
    "Valor": [10, 25, 15, 30, 18]
}).sort_values("Valor")

plt.figure(figsize=(7, 4))

# “haste” do pirulito
for _, row in df.iterrows():
    plt.plot([0, row["Valor"]], [row["Categoria"], row["Categoria"]], linewidth=2)

# “cabeça”
plt.scatter(df["Valor"], df["Categoria"], s=60)

plt.title("Lollipop: mesma ideia do dot plot com referência")
plt.xlabel("Valor")
plt.ylabel("Categoria")
plt.tight_layout()
plt.show()
```

# Small multiples (por subgrupo)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Regiao": ["Norte","Norte","Sul","Sul","Leste","Leste"],
    "Categoria": ["A","B","A","B","A","B"],
    "Vendas": [40, 55, 70, 45, 60, 65]
})

regioes = df["Regiao"].unique()

fig, axes = plt.subplots(1, len(regioes), figsize=(12, 3), sharey=True)

for ax, reg in zip(axes, regioes):
    dfr = df[df["Regiao"] == reg].sort_values("Vendas", ascending=False)
    ax.barh(dfr["Categoria"], dfr["Vendas"])
    ax.set_title(reg)
    ax.invert_yaxis()
    ax.set_xlabel("Vendas")

axes[0].set_ylabel("Categoria")
plt.tight_layout()
plt.show()
```
