```python
import seaborn as sns
import matplotlib.pyplot as plt

# carregar dataset
df = sns.load_dataset("tips")
```

```python
df
```

```python
plt.figure(figsize=(16,5))

# ---------------------
# BOX PLOT
# ---------------------
plt.subplot(1,3,1)
sns.boxplot(
    data=df,
    x="day",
    y="total_bill",
    palette="Blues"
)

plt.title("Boxplot\nResumo estatístico e outliers")
plt.xlabel("Dia")
plt.ylabel("Conta total")

# ---------------------
# VIOLIN PLOT
# ---------------------
plt.subplot(1,3,2)

sns.violinplot(
    data=df,
    x="day",
    y="total_bill",
    palette="Blues"
)

plt.title("Violin Plot\nForma da distribuição")
plt.xlabel("Dia")
plt.ylabel("Conta total")

# ---------------------
# DENSIDADE
# ---------------------
plt.subplot(1,3,3)

sns.kdeplot(
    data=df,
    x="total_bill",
    hue="day",
    fill=True,
    alpha=0.4
)

plt.title("Densidade\nComparação contínua")
plt.xlabel("Conta total")

plt.tight_layout()
plt.show()
```
