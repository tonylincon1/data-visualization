```python
import numpy as np
import matplotlib.pyplot as plt

# reprodutibilidade
np.random.seed(42)

# dados simulados: maioria concentrada + alguns valores extremos
dados = np.random.normal(loc=50, scale=10, size=500)
outliers = np.array([5, 10, 95, 100])
dados_com_outliers = np.concatenate([dados, outliers])
```

```python
# -----------------------------
# 1) Histograma com bins "intermediários"
# -----------------------------
plt.figure(figsize=(8, 4))
plt.hist(dados_com_outliers, bins=15, edgecolor="black")
plt.title("Histograma - distribuição dos dados")
plt.xlabel("Valor")
plt.ylabel("Frequência")
plt.show()
```

```python
# -----------------------------
# 2) Poucos bins
# -----------------------------
plt.figure(figsize=(8, 4))
plt.hist(dados_com_outliers, bins=5, edgecolor="black")
plt.title("Histograma com poucos bins")
plt.xlabel("Valor")
plt.ylabel("Frequência")
plt.show()
```

```python
# -----------------------------
# 3) Muitos bins
# -----------------------------
plt.figure(figsize=(8, 4))
plt.hist(dados_com_outliers, bins=40, edgecolor="black")
plt.title("Histograma com muitos bins")
plt.xlabel("Valor")
plt.ylabel("Frequência")
plt.show()
```
