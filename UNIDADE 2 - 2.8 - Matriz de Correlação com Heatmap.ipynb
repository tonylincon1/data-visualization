```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

n = 200

# gerar variáveis com relações moderadas e ruído
horas_estudo = np.random.normal(5, 1.5, n)
horas_sono = np.random.normal(7, 1, n)
redes_sociais = np.random.normal(3, 1.5, n)

nota = (
    horas_estudo * 0.6
    - redes_sociais * 0.3
    + np.random.normal(0, 1.5, n)
)

estresse = (
    redes_sociais * 0.5
    - horas_sono * 0.4
    + np.random.normal(0, 1.2, n)
)

df = pd.DataFrame({
    "Horas de Estudo": horas_estudo,
    "Nota na Prova": nota,
    "Horas de Sono": horas_sono,
    "Uso de Redes Sociais": redes_sociais,
    "Nível de Estresse": estresse
})
```

```python
df
```

```python
corr = df.corr()

# mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(9,6))

sns.heatmap(
    corr,
    # mask=mask,
    annot=True,
    fmt=".2f",
    cmap="RdBu",
    vmin=-1,
    vmax=1,
    center=0,
    linewidths=1,
    square=True,
    annot_kws={"size":12}
)

plt.title("Correlação entre Hábitos e Desempenho Acadêmico", fontsize=16)

plt.xticks(rotation=30, ha="right", fontsize=11)
plt.yticks(rotation=0, fontsize=11)

plt.tight_layout()
plt.show()
```
