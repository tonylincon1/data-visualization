```python
import pandas as pd
import matplotlib.pyplot as plt

# criando dados diários (3 meses)
df = pd.DataFrame({
    "data": pd.date_range("2025-01-01", periods=90, freq="D")
})
df["vendas"] = (100 + (df.index * 0.5) + (df.index % 7) * 3).astype(float)  # tendência + variação simples
```

```python
df
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df["data"], df["vendas"], linewidth=1, label="Diário")
plt.title("Série temporal")
plt.xlabel("Data")
plt.ylabel("Vendas")
plt.legend()
plt.tight_layout()
plt.show()
```
